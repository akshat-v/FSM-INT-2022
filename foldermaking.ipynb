```python
for filename in glob('/content/drive/MyDrive/PCB dataset/1/all_fault/*.jpg'):
 
  root_img_name=filename.replace('all_fault/','all_temp/')
  train_name = root_img_name.replace('_test.jpg', '')
  train_name= train_name + '_temp.jpg'
  dest_name= filename.replace('all_fault/','all_mask/')
  dest_name=dest_name.replace('_test.jpg','_mask.jpg')
  image1 = cv2.imread(filename,0)
  image2 = cv2.imread(train_name,0)
  bitwise_xor = cv2.bitwise_xor(image1, image2)
  new_image = cv2.medianBlur(bitwise_xor, 7)

  
  cv2.imwrite(dest_name, new_image)
```
